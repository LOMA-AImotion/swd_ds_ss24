Wir ziehen zunächst 1000 zufällige Werte aus einer Normalverteilung mit benutzerdefiniertem Mittelwert und Varianz-Parameter und visualisieren dann das Ergebnis. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Benutzerdefinierte Parameter für die Normalverteilung
mittelwert = 180  # Mittelwert (mean)
varianz = 25      # Varianz (variance)
standardabweichung = np.sqrt(varianz)  # Standardabweichung (standard deviation)

# 1000 zufällige Werte aus der Normalverteilung ziehen
zufallswerte = np.random.normal(mittelwert, standardabweichung, 1000)

# TODO: Geben Sie hier die ersten 10 Werte aus zufallswerte aus (Hinweis: Slicing!)
# print(...)
```

Geben Sie nun ein Histogramm der zufälligen Werte aus. (Hinweis `plt.hist`, hier können Sie gerne KI befragen)

```python
# Histogramm der zufälligen Werte
plt.hist(zufallswerte, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogramm der zufälligen Werte aus einer Normalverteilung')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.grid(True)
plt.show()
```

Geben Sie nun in dem Histogramm auch die Parameter `mittelwert` und `standardabweichung` aus:

```python
# Histogramm der zufälligen Werte
plt.hist(zufallswerte, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Histogramm der zufälligen Werte aus einer Normalverteilung')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.grid(True)

# Mittelwert und Varianzlinien hinzufügen
plt.axvline(mittelwert, color='red', linestyle='dashed', linewidth=2, label='Mittelwert')
plt.axvline(mittelwert - standardabweichung, color='green', linestyle='dashed', linewidth=2, label='-1 Std. Abw.')
plt.axvline(mittelwert + standardabweichung, color='green', linestyle='dashed', linewidth=2, label='+1 Std. Abw.')

# Plot Titel und Achsenbeschriftungen
plt.title('Histogramm der zufälligen Werte aus einer Normalverteilung')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.grid(True)
plt.legend()
plt.show()
```
