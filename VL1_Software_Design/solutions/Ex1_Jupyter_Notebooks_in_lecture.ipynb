Wir ziehen zunächst 1000 **zufällige** Werte aus einer Normalverteilung mit benutzerdefiniertem Mittelwert und Varianz-Parameter und visualisieren dann das Ergebnis. 

```python
import numpy as np
import matplotlib.pyplot as plt

def ziehe_stichprobe(mittelwert, standardabweichung, anzahl, seed=0):
    # 1000 zufällige Werte aus der Normalverteilung ziehen
    np.random.seed(seed)
    zufallswerte = np.random.normal(mittelwert, standardabweichung, anzahl)
    return zufallswerte

# Benutzerdefinierte Parameter für die Normalverteilung
mittelwert = 180  # Mittelwert (mean)
varianz = 25      # Varianz (variance)
standardabweichung = np.sqrt(varianz)  # Standardabweichung (standard deviation)

zufallswerte = ziehe_stichprobe(mittelwert, standardabweichung, 1000)
# TODO: Geben Sie hier die ersten 10 Werte aus zufallswerte aus (Hinweis: Slicing!)
print(zufallswerte[0:11])
```

Geben Sie nun ein Histogramm der zufälligen Werte aus. (Hinweis `plt.hist`, hier können Sie gerne KI befragen)

```python
def plot_zufallswerte(zufallswerte):
    # Erstellt das Histogramm.
    plt.hist(zufallswerte, bins=50, edgecolor='black')
    # Zeichnet eine vertikale Linie bei mittelwert Wert.
    plt.axvline(mittelwert, color='r', linestyle='dashed', linewidth=2, label = 'Mittelwert')
    # Zeichnet eine vertikale Linie bei standardabweichung Wert.
    plt.axvline(mittelwert + standardabweichung, color='g', linestyle='dashed', linewidth=2, label = 'Standardabweichung')
    plt.axvline(mittelwert - standardabweichung, color='g', linestyle='dashed', linewidth=2)

    # Sie können hier Titel, x-Achsenbeschriftung und y-Achsenbeschriftung hinzufügen.
    plt.title("Histogramm der Zufallswerte")
    plt.xlabel("Wert")
    plt.ylabel("Häufigkeit")
```

Geben Sie nun in dem Histogramm auch die Parameter `mittelwert` und `standardabweichung` aus (Hinweis `plt.axvline`):

```python
plot_zufallswerte(zufallswerte)
```

Nun *wiederholen* Sie das Experiment mit zwei weiteren Random-Seeds (z.B. 42 und 1337)

```python
for seed in [42, 1337]:
    zufallswerte = ziehe_stichprobe(mittelwert, standardabweichung, 1000, seed)
    plot_zufallswerte(zufallswerte)
    plt.figure()
```
