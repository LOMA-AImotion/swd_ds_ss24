```python
import pandas as pd
import numpy as np
from scipy.stats import linregress

import os

data_file = 'data/women_in_parliament_processed.csv'
df = pd.read_csv(data_file)
timestamps = [int(i) for i in df.index.tolist()]
de_parl = df["Germany"].tolist()
years = df["Unnamed: 0"].tolist()

def fit_trendline(year_timestamps, data):
    result = linregress(year_timestamps, data)
    slope = round(result.slope, 3)
    intercept = round(result.intercept, 3)
    r_value = round(result.rvalue**2, 3)
    return slope, intercept, r_value

slope, intercept, r_value = fit_trendline(timestamps, de_parl)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(years, de_parl, 'rx')
plt.ylim(30, 38)
plt.title("Women percentage in parliaments vs. years")
plt.xlabel("Years")
plt.ylabel("Women in parliaments perc.")
# Calculate the predicted values
trendline = [slope * t + intercept for t in timestamps]
plt.plot(years, de_parl, 'rx', label='Actual Data')
plt.plot(years, trendline, 'b-', label=f'Trendline (slope={slope}, $R$={r_value})')
plt.legend()

```
